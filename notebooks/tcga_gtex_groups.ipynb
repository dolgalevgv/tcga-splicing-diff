```python
def assign_prevailing_forms(dataset: pd.DataFrame) -> pd.DataFrame:
    ''' 
    Given initial TCGA dataset, iterate through unique genes and:
    1. If gene has a single isoform in dataset, simply assign it to all samples
    2. If gene has multiple isoforms, find isoforms with maximum abundance for
    each sample using .idxmax() method and assign the result to samples, 
    collapsing data to one row per gene.
    '''
    #Construct a dataframe to return from initial dataset, drop unnecessary columns
    result = (dataset.copy()
              .drop('uniprot_isoform', axis=1)
              .drop_duplicates('ensembl_gene_id')
              .set_index('ensembl_gene_id'))
    
    #Iterate over genes
    for i, g in enumerate(result.index):
        subset_g = dataset[dataset['ensembl_gene_id'] == g]
        subset_g = subset_g.set_index('uniprot_isoform', drop=True).iloc[:,1:]
        result.iloc[i] = subset_g.idxmax()
            
    return result
```

```python
def extract_dominant_form(isoform_list: pd.Series):
    ''' 
    Find frequencies of prevailing isoforms for a given gene
    Return a dominant isoform if its frequency >= 80%
    Else return NaN
    '''
    isoform_counts = isoform_list.value_counts(normalize=True)
    if isoform_counts[0] >= 0.8: # index 0 since value_counts auto sorts results
        return isoform_counts.index[0]
    else:
        return np.nan
```

```python
start = timeit.default_timer()

consensus_canc = pd.DataFrame(index=toil['ensembl_gene_id'].drop_duplicates(), columns=samples)
consensus_norm = pd.DataFrame(index=toil['ensembl_gene_id'].drop_duplicates(), columns=samples)

for i, s in enumerate(meta_tcga_normal.index):
    new = timeit.default_timer()
    
    sc_samples = meta_tcga[(meta_tcga['detailed_category'] == s) & (meta_tcga['_sample_type'] == 'Primary Tumor')]['sample'].to_list()
    sc_data = toil[['uniprot_isoform', 'ensembl_gene_id'] + sc_samples]
    sc_prev = assign_prevailing_forms(sc_data)
    consensus_canc[f'{tcga_studies.loc[s].values[0]}'] = consensus_canc.index.map(sc_prev.apply(extract_dominant_form, axis=1))
      
    sn_samples = meta_tcga[(meta_tcga['detailed_category'] == s) & (meta_tcga['_sample_type'] == 'Solid Tissue Normal')]['sample'].to_list()
    sn_data = toil[['uniprot_isoform', 'ensembl_gene_id'] + sn_samples]
    sn_prev = assign_prevailing_forms(sn_data)
    consensus_norm[f'{tcga_studies.loc[s].values[0]}'] = consensus_norm.index.map(sn_prev.apply(extract_dominant_form, axis=1))
    
    elapsed_one = timeit.default_timer() - new
    elapsed_total = timeit.default_timer() - start
    
    msg = f'Done {i+1} out of {len(samples)} samples, last took {elapsed_one:.2f} s, total {elapsed_total:.2f} s          '
    
    print(msg, end='\r')
```
